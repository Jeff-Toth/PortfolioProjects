```python
#Clean up data
```

```python
#Only looking at USA Races, 50k or 50mi, 2020
```

```python
#Step 1: show 50mi and and 50k
#50km
#50mi
```

```python
#Checking notation for miles

df[df['Event distance/length'] == '50mi']
```

```python
#combining 50k/50mi with isin

df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
#filtering for year 2020

df[(df['Event distance/length'].isin(['50km', '50mi']))&(df['Year of event'] == 2020)]
```

```python
df[df['Event name'] == 'Yankee Springs 50 Mile Winter Challenge (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#Combining all of the filters together

df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
#remove (USA) from event name column
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Cleaning up athlete age

df2['Athlete age'] = 2020 - df['Athlete year of birth']
```

```python
#Removing h from athlete performance

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category

df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
df2.shape
```

```python
#Looking at null values

df2.isna().sum()
```

```python
df2[df2['Athlete age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#checking for duplicates
df2[df2.duplicated() == True]
```

```python
#reset index for updtated dataframce

df2.reset_index(drop = True)
```

```python
#Fixing types

df2.dtypes
```

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#rename columns for convenience sake
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#Athlete age                    int32
```

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_date',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_number_of_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_average_speed',
                            'Athlete ID': 'athlete_id',
                            'Athlete age': 'athlete_age'
                             })
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3 = df2[['race_date', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed', 'year']]
```

```python
df3.head()
```

```python
#chart and graphs
```

```python
# comparing 50km vs. 50mi

sns.histplot(df3['race_length'])

plt.title('50km vs. 50mi')
plt.xlabel('Race Length')
plt.ylabel('Count')
plt.show()
```

```python
#comparision by gender

sns.histplot(df3, x = 'race_length', hue = 'athlete_gender')

plt.title('50km vs. 50mi by Gender')
plt.xlabel('Race Length')
plt.ylabel('Count')
plt.show()
```

```python
#50 mile average speed distribution

sns.histplot(df3[df3['race_length'] == '50mi']['athlete_average_speed'])

plt.title('Average Speed for 50 mile')
plt.xlabel('Average Speed')
plt.ylabel('Count')
plt.show()
```

```python
#violin

sns.violinplot(data = df3, x='race_length', y='athlete_average_speed', hue='athlete_gender', split=True, inner='quart', linewidth=1)
```

```python
#
sns.lmplot(data=df3, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
```

```python
#comparing average speeds by season

#winter 12 - 2
#spring 3 - 5
#summer 6 - 8
#fall 9 - 11
```

```python
df3['race_month'] = df3['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head(25)
```

```python
#All races
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#50 mile races
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#50km races
df3.query('race_length == "50km"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
